**Q2:** Aik `BankAccount` class banayein jismein `__balance` private variable ho. Aik `deposit()` aur aik `withdraw()` method banayein, check karein ke withdraw karte waqt balance zero se kam na ho.

```python
class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive")
    
    def withdraw(self, amount):
        if amount > 0:
            if self.__balance - amount >= 0:
                self.__balance -= amount
                print(f"Withdrawn: ${amount}. New balance: ${self.__balance}")
            else:
                print("Insufficient balance")
        else:
            print("Withdraw amount must be positive")
    
    def get_balance(self):
        return self.__balance
```

```python
account = BankAccount(1000)
```

```python
account.deposit(500)
```

```python
account.withdraw(200) 
```

```python
account.withdraw(2000)
```

**Q9:** Child class ke constructor (`__init__`) ke andar parent class ke constructor ko call karne ke liye `super()` ka use kaise hota hai? Aik practical example likhein.

```python
class Parent:
    def __init__(self, name):
        self.name = name
        print(f"Parent constructor called with name: {name}")
    
    def display(self):
        print(f"Parent name: {self.name}")

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age
        print(f"Child constructor called with age: {age}")
    
    def display_all(self):
        super().display()
        print(f"Child age: {self.age}")
```

```python
child_obj = Child("shakir", 10)
child_obj.display_all()
```

**Q10:** Hierarchical Inheritance kya hoti hai? Aik parent class `Shape` banayein aur us se do alag child classes `Circle` aur `Square` banakar samjhayein.

```python
class Shape:
    def __init__(self, color="red"):
        self.color = color
    
    def get_color(self):
        return self.color

class Circle(Shape):
    def __init__(self, radius, color="red"):
        super().__init__(color)
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2
    
    def display(self):
        print(f"Circle with radius {self.radius}, color {self.color}")

class Square(Shape):
    def __init__(self, side, color="red"):
        super().__init__(color)
        self.side = side
    
    def area(self):
        return self.side ** 2
    
    def display(self):
        print(f"Square with side {self.side}, color {self.color}")
```

```python
circle = Circle(5, "blue")
square = Square(4, "green")
```

```python
circle.display()
square.display() 
```

```python
print(f"Circle area: {circle.area()}")
print(f"Square area: {square.area()}")
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Point(new_x, new_y)
    
    def display(self):
        print(f"Point({self.x}, {self.y})")
```

```python
p1 = Point(2, 3)
p2 = Point(4, 5)
p3 = p1 + p2
```

```python
p1.display()
p2.display()  
p3.display() 
```

**Q15:** Kya Python default roop se Method Overloading (same method name with different arguments) support karta hai? Agar nahi, toh hum same name ke method ko multiple arguments ke sath kaise handle karte hain? Code likhein.

```python
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c
class Calculator2:
    def add(self, *args):
        return sum(args)

class Calculator3:
    def multiply(self, *args):
        if len(args) == 1:
            return args[0] ** 2
        elif len(args) == 2:
            return args[0] * args[1]
        else:
            result = 1
            for num in args:
                result *= num
            return result
```

```python
calc = Calculator()
print(calc.add(5))          
print(calc.add(5, 10)) 
print(calc.add(5, 10, 15)) 
```

```python
calc2 = Calculator2()
print(calc2.add(1, 2, 3, 4, 5)) 
```

```python
calc3 = Calculator3()
print(calc3.multiply(5))          
print(calc3.multiply(5, 6))     
print(calc3.multiply(2, 3, 4))
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass
    
    @abstractmethod
    def get_status(self):
        pass

class CreditCardPayment(Payment):
    def __init__(self, card_number, expiry, cvv):
        self.__card_number = card_number
        self.__expiry = expiry
        self.__cvv = cvv
        self.__status = "Pending"
    
    def process_payment(self, amount):
        print(f"Processing Credit Card payment of ${amount}")
        print(f"Card: ****{self.__card_number[-4:]}")
        self.__status = "Completed"
        return True
    
    def get_status(self):
        return self.__status

class UPIPayment(Payment):
    def __init__(self, upi_id):
        self.__upi_id = upi_id
        self.__status = "Pending"
    
    def process_payment(self, amount):
        print(f"Processing UPI payment of ${amount}")
        print(f"UPI ID: {self.__upi_id}")
        self.__status = "Completed"
        return True
    
    def get_status(self):
        return self.__status

class PayPalPayment(Payment):
    def __init__(self, email):
        self.__email = email
        self.__status = "Pending"
    
    def process_payment(self, amount):
        print(f"Processing PayPal payment of ${amount}")
        print(f"Email: {self.__email}")
        self.__status = "Completed"
        return True
    
    def get_status(self):
        return self.__status

def make_payment(payment_method, amount):
    payment_method.process_payment(amount)
    print(f"Payment Status: {payment_method.get_status()}\n")

```

```python
credit_card = CreditCardPayment("1234567890", "12/25", "123")
upi = UPIPayment("user@abcbank")
paypal = PayPalPayment("abc@gmail.com")
```

```python
make_payment(credit_card, 5000)
make_payment(upi, 2500)
make_payment(paypal, 7500)
```
